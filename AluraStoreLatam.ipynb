#1. Análisis de facturación



```python
#Analisis
#concateno para lograr un mejor analisis (todos juntos)

tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)

#Facilita:

#calcular ingresos totales

#ver categorías más vendidas

#analizar reseñas

#sacar promedios
```

```python
tiendas.head()

```

```python
tiendas.info()
```

```python
#LIMPIEZA DE NULOS
tiendas.isnull().sum()
```

```python
#Normalizar Nombres
tiendas.columns = (
    tiendas.columns
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("í", "i")
    .str.replace("ó", "o")
)
```

```python
#Calculo de ingresos

tiendas["ingreso"] = tiendas["precio"]

```

```python
ingreso_total = tiendas["ingreso"].sum()
ingreso_total
print(f"Ingreso total: {ingreso_total:,.0f}")

```

```python
#Productos mas Vendidos
productos_mas_vendidos = (
    tiendas["producto"]
    .value_counts()
)

productos_mas_vendidos.head(10)
```

# 3. Calificación promedio de la tienda


```python
#CALIFICACION
calificacion_promedio = tiendas["calificacion"].mean()
calificacion_promedio
print(float(calificacion_promedio))
```

```python
calificacion_por_categoria = (
    tiendas.groupby("categoria_del_producto")["calificacion"]
    .mean()
    .sort_values(ascending=False)
)

calificacion_por_categoria
```

# 5. Envío promedio por tienda

```python
def normalizar_columnas(df):
    df.columns = (
        df.columns
        .str.lower()
        .str.replace(" ", "_")
        .str.replace("í", "i")
        .str.replace("ó", "o")
    )
    return df

tienda1 = normalizar_columnas(tienda1)
tienda2 = normalizar_columnas(tienda2)
tienda3 = normalizar_columnas(tienda3)
tienda4 = normalizar_columnas(tienda4)
```

```python
envio_tienda1 = tienda1["costo_de_envio"].mean()
envio_tienda2 = tienda2["costo_de_envio"].mean()
envio_tienda3 = tienda3["costo_de_envio"].mean()
envio_tienda4 = tienda4["costo_de_envio"].mean()

envio_tienda1, envio_tienda2, envio_tienda3, envio_tienda4
```

```python
envio_promedio_por_tienda = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Costo de envío promedio": [
        envio_tienda1,
        envio_tienda2,
        envio_tienda3,
        envio_tienda4
    ]
})

envio_promedio_por_tienda
```

```python
envio_promedio_por_tienda.set_index("Tienda").plot(
    kind="pie",
    y="Costo de envío promedio",
    autopct="%1.1f%%",
    title="Distribución del costo de envío promedio por tienda",
    legend=False
)
```

```python
envio_promedio_por_tienda.plot(
    x="Tienda",
    y="Costo de envío promedio",
    kind="bar",
    title="Costo de envío promedio por tienda",
    legend=False,
    xlabel="Tienda",
    ylabel="Costo de envío promedio"
)
```
